````python
# We import tensorflow as tf because tf is easyer to write
import tensorflow as tf
````

````python
#Import dataset using Tensorflow

mnist = tf.keras.datasets.mnist
#Separating the data between the 'TrainSet' and the 'TestSet'
(x_train,y_train),(x_test,y_test) = mnist.load_data()
````

````python
#Using the normalize function helps to prevent loss of information during training and also ensures that the model can generalize well on unseen data.
x_train = tf.keras.utils.normalize(x_train, axis=1)
x_test = tf.keras.utils.normalize(x_test, axis=1)
````

````python
#Choosing the models "Sequential()"
model = tf.keras.models.Sequential()
````

````python
# Here we used the"layers Flatten" to put the image who is a 2d vector into a 1d vector from (28x28) to (1x784)
# This Layer is the entry one


model.add(tf.keras.layers.Flatten(input_shape=(28,28)))
#Here is the building the two "hiden layers" Who will ajst the wheigt and biais of the neural network
model.add(tf.keras.layers.Dense(128, activation='relu'))
model.add(tf.keras.layers.Dense(128, activation='relu'))
#Here is the output of the neural network, There is only 10 possible output so the output number is 10.
model.add(tf.keras.layers.Dense(10, activation='softmax'))
````

````python
#Here is the compile fonction where we define 
# Why Adam ? This is what i found in the tensorflow docs
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
````

````python
#Fiting the model using the fit() function
model.fit(x_train, y_train, epochs=3)
#Adding epochs can help improving the accuracy
model_5 = model.fit(x_train, y_train, epochs=5)
````

## Output analysis

We can check the mean of the 2 epochs :
- The model_3 with 3 epoch have a accuracy of 95,54% and a loss of 0,1480
- The model_5 with 5 epoch have a accuracy of 98,88% and a loss of 0,138

We can see that the at the 3/3 epochs the accuracy was at 97,77% and at the 5/5 epochs the acuracy was at 99,26%
